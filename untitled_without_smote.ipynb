```python
import numpy as np
import pandas as pd
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score, confusion_matrix, recall_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("creditcard.csv")
print(df.shape)
```

```python
df.head()
```

```python
p_class = df[df.Class == 1]
n_class = df[df.Class == 0]

print(p_class.shape, n_class.shape)
```

```python
ne_class = n_class.sample(n=492)
print(ne_class.shape)
```

```python
new_set = pd.concat([ne_class, p_class], axis=0)
print(new_set.shape)
```

```python
new_set.head()
```

```python
X = new_set.drop("Class", axis=1)
y = new_set["Class"]

print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=2)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
model = XGBClassifier(
    n_estimators = 200,
    max_depth = 6,
    scale_pos_weight = 10,
    eval_metric = 'logloss'
)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
help(confusion_matrix)
```

```python
help(roc_auc_score)
```

```python
y_pred_proba = model.predict_proba(X_test)[:, 1]

roc = roc_auc_score(y_test, y_pred_proba)
print(f"roc_auc score is: {roc}")
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
rc = recall_score(y_test, y_pred)
print(f"Recall score is: {rc}")
```
